```python
try:
    # Initialize the browser
    driver = webdriver.Chrome()

    # Navigate to the URL
    driver.get('https://www.bikedekho.com')

    # Example search queries list
    search_queries = ['Hero bike', 'Bajaj bike', 'TVS bike', 'Royal Enfield bike', 'Honda bike', 'Yamaha bike', 'Suzuki bike', 'KTM bike', 'Jawa bike', 'Benelli bike', 'Harley bike', 'Triumph bike', 'Ducati bike', 'BMW bike', 'Kawasaki bike', 'Husqvarna bike', 'Revolt bike', 'Ather bike']
    # search_queries = ['ola bike']

    for search_query in search_queries:
        # Find the search text input element
        search_text_input = driver.find_element(By.ID, 'cardekhosearchtext')

        # Input the search query into the search text input
        search_text_input.clear()  # Clear any previous input
        search_text_input.send_keys(search_query)

        # Wait for a while to let the dropdown options appear
        sleep(2)

        # Re-find the dropdown options after waiting
        dropdown_option_xpath = '//*[@id="rf01"]/header/div[1]/div/div/div[2]/div/div/form/div/div/div/ul/li'
        dropdown_options = driver.find_elements(By.XPATH, dropdown_option_xpath)

        # Select the first option from the dropdown if available
        if dropdown_options:
            first_option = dropdown_options[0]
            first_option.click()

        sleep(2)

        bike_list_element = driver.find_element(By.CLASS_NAME, 'bikelist')

        bike_details =[]

        # Extract data from the bike list
        for index, item in enumerate(bike_list_element.find_elements(By.TAG_NAME, 'li')):
            # Skip list items 6 and 7
            if index == 5 or index == 6:
                continue

            # Extract data from each item as needed
            bike_name = item.find_element(By.TAG_NAME, 'h3').text
            bike_price = item.find_element(By.CLASS_NAME, 'price').text
            basic_specs = item.find_element(By.CLASS_NAME, 'dotlist').text

            bike_df = {'Name': bike_name, 'Price': bike_price, 'Basic Specs': basic_specs}

            bike_details.append(bike_df)
            

        # bike_details = []
        # for item_index, item in enumerate(bike_items):
        #     # Skip list item 6 and 7
        #     if item_index == 5 or item_index == 6:
        #         continue

        #     bike_name = item.find_element(By.TAG_NAME, 'h3').text
        #     bike_price = item.find_element(By.CLASS_NAME, 'price').text
        #     basic_specs = item.find_element(By.CLASS_NAME, 'dotlist').text

        #     bike_details.append({'Name': bike_name, 'Price': bike_price, 'Basic Specs': basic_specs})

        # # Print scraped data for the brand
        # print(f"Scraped data for Brand {index + 1}: {brand_name}")
        
        
        # Save scraped data to a DataFrame
        bikes_df_basics1 = pd.DataFrame(bike_details)
        # bikes_df_basics1.to_csv(f"{bike_name}_bikes.csv", index=False)  # Save data to CSV file
        print(bikes_df_basics1)

except Exception as e:
    print("An error occurred:", str(e))


finally:
    # Close the browser
    driver.quit()
```

```python
import re

# List of bike brands in India
bike_brands_in_india = [
    "Hero MotoCorp",
    "Bajaj Auto",
    "TVS Motor Company",
    "Royal Enfield",
    "Honda Motorcycle and Scooter India (HMSI)",
    "Yamaha Motor India",
    "Suzuki Motorcycle India Private Limited (SMIPL)",
    "KTM India",
    "Mahindra Two Wheelers",
    "Jawa Motorcycles",
    "Benelli India",
    "Harley-Davidson India",
    "Triumph Motorcycles India",
    "Ducati India",
    "BMW Motorrad India",
    "Kawasaki Motors Pvt. Ltd.",
    "Husqvarna Motorcycles India",
    "Revolt Motors",
    "Ather Energy"
]

# Function to add "bike" text after the first word and remove everything after that addition
def add_bike_and_remove_after(lst):
    updated_list = []
    for brand in lst:
        # Using regular expression to find the first word
        first_word_match = re.match(r'\w+', brand)
        if first_word_match:
            first_word = first_word_match.group()
            # Adding "bike" after the first word
            updated_brand = first_word + " bike"
            updated_list.append(updated_brand)
    return updated_list

# Apply the function to the list of bike brands
updated_bike_brands = add_bike_and_remove_after(bike_brands_in_india)

# Print the updated list of bike brands
print(updated_bike_brands)
```
